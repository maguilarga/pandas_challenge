# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Differences in % Overall Passing Rate. Wrong Calculation, it is NOT a rate  ********

# Calculate the different data points
tot_students = student_data["Student ID"].count()
# passing math defined as Number of students with math score >= 70
pass_math = student_data.query("math_score >= 70")["Student ID"].count()
# Passing reading defined as Number of students with reading score >= 70
pass_read = student_data.query("reading_score >= 70")["Student ID"].count()
# Overall score defined as the average of the addition of math and reading scores
# Passing overall defined as Overall score >= 70
pass_over = student_data[["Student ID"]].\
    assign(overall_score=lambda x: (student_data['math_score'] + student_data['reading_score']) / 2).\
    query("overall_score >= 70")["Student ID"].count()

# Create the data frame with the required data
district_summary = pd.DataFrame ({
    "Total Schools": [school_data["School ID"].count()],
    "Total Students": tot_students,
    "Total Budget": [school_data["budget"].sum()],
    "Average Math Score": [student_data["math_score"].mean()],
    "Average Reading Score": [student_data["reading_score"].mean()],
    "Overall Average Score": [(student_data["math_score"].mean()+student_data["reading_score"].mean())/2],
    "% Passing Math": [(pass_math/tot_students) * 100],
    "% Passing Reading": [(pass_read/tot_students) * 100],
    "% Overall Passing Rate": [(pass_over/tot_students) * 100]
#   Another way of calculating it
#   "% Overall Passing Rate": [((pass_math/tot_students) + (pass_read/tot_students) )/2] 
        
})

# Format Data
district_summary["Total Schools"] = district_summary["Total Schools"].map("{:,}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,.0f}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["Overall Average Score"] = district_summary["Overall Average Score"].map("{:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2f}%".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map("{:,.2f}%".format)

district_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# freate the data frame with the requested data
school_summary = school_data_complete.groupby('school_name')\
    .agg({'type': 'max', "size": 'mean', 'budget': 'mean', 'math_score': 'mean', 'reading_score': 'mean'})\
    .rename(columns={'type': 'School Type', 'size': 'Total Students', 'budget': 'Total School Budget', \
                    'math_score': 'Average Math Score', 'reading_score': 'Average Reading Score'})
school_summary["Per Student Budget"] = school_summary["Total School Budget"] / school_summary["Total Students"]
school_summary["Overall Average Score"] = (school_summary['Average Math Score'] + school_summary['Average Reading Score']) / 2

# calculate the counts of people for passing math and reading to be able to calculate percentages
school_summary["CT_Passing_Math"] = school_data_complete.query("math_score >= 70")[['school_name','math_score']]\
    .groupby('school_name').agg({'math_score': 'count'})
school_summary["CT_Passing_Reading"] = school_data_complete.query("reading_score >= 70")[['school_name','reading_score']]\
    .groupby('school_name').agg({'reading_score': 'count'})

# Calculate the percentage of students passed in math and reading
school_summary["% Passing Math"] = (school_summary['CT_Passing_Math'] / school_summary['Total Students']) * 100
school_summary["% Passing Reading"] = (school_summary['CT_Passing_Reading'] / school_summary['Total Students']) * 100
# Using the easy calculation for % overall: average of %math and %reading
school_summary["% Overall Passing Rate"] = (school_summary["% Passing Math"] + school_summary["% Passing Reading"]) / 2

# Reordering columns to the stated output and dropping columns that are NOT required
school_summary = school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",\
    "Average Math Score", "Average Reading Score", "Overall Average Score", "% Passing Math", "% Passing Reading",\
    "% Overall Passing Rate"]]

# Sort to show top performers and Format Data
top_perf = school_summary.sort_values("% Overall Passing Rate", ascending=False)
top_perf["Total Students"] = top_perf["Total Students"].map("{:,.0f}".format)
top_perf["Total School Budget"] = top_perf["Total School Budget"].map("${:,.2f}".format)
top_perf["Per Student Budget"] = top_perf["Per Student Budget"].map("${:,.2f}".format)
top_perf["Average Math Score"] = top_perf["Average Math Score"].map("{:,.2f}".format)
top_perf["Average Reading Score"] = top_perf["Average Reading Score"].map("{:,.2f}".format)
top_perf["Overall Average Score"] = top_perf["Overall Average Score"].map("{:,.2f}".format)
top_perf["% Passing Math"] = top_perf["% Passing Math"].map("{:,.2f}%".format)
top_perf["% Passing Reading"] = top_perf["% Passing Reading"].map("{:,.2f}%".format)
top_perf["% Overall Passing Rate"] = top_perf["% Overall Passing Rate"].map("{:,.2f}%".format)

top_perf.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
# Sort to show bottom performers and Format Data
botom_perf = school_summary.sort_values("% Overall Passing Rate", ascending=True)
botom_perf["Total Students"] = botom_perf["Total Students"].map("{:,.0f}".format)
botom_perf["Total School Budget"] = botom_perf["Total School Budget"].map("${:,.2f}".format)
botom_perf["Per Student Budget"] = botom_perf["Per Student Budget"].map("${:,.2f}".format)
botom_perf["Average Math Score"] = botom_perf["Average Math Score"].map("{:,.2f}".format)
botom_perf["Average Reading Score"] = botom_perf["Average Reading Score"].map("{:,.2f}".format)
botom_perf["Overall Average Score"] = botom_perf["Overall Average Score"].map("{:,.2f}".format)
botom_perf["% Passing Math"] = botom_perf["% Passing Math"].map("{:,.2f}%".format)
botom_perf["% Passing Reading"] = botom_perf["% Passing Reading"].map("{:,.2f}%".format)
botom_perf["% Overall Passing Rate"] = botom_perf["% Overall Passing Rate"].map("{:,.2f}%".format)

botom_perf.head(5)
```

```python
# Alternative way of getting "Math Scores by Grade". The grades are in differnt order
math_grade_alt = school_data_complete.groupby(["school_name","grade"]).agg({"math_score": "mean"}).unstack("grade")

# Using pivot
math_grade_alt = school_data_complete.groupby(["school_name","grade"]).agg({"math_score": "mean"}).reset_index()\
    .pivot(values='math_score', index='school_name', columns='grade')

math_grade_alt
```

```python
# calculate math avg scores by school for each grade
math9 = school_data_complete.loc[(school_data_complete["grade"] == '9th'),["school_name", "math_score"]]\
    .groupby("school_name").agg({"math_score": "mean"})
math10 = school_data_complete.loc[(school_data_complete["grade"] == '10th'),["school_name", "math_score"]]\
    .groupby("school_name").agg({"math_score": "mean"})
math11 = school_data_complete.loc[(school_data_complete["grade"] == '11th'),["school_name", "math_score"]]\
    .groupby("school_name").agg({"math_score": "mean"})
math12 = school_data_complete.loc[(school_data_complete["grade"] == '12th'),["school_name", "math_score"]]\
    .groupby("school_name").agg({"math_score": "mean"})

math9 = pd.Series(math9.math_score, index=math9.index)
math10 = pd.Series(math10.math_score, index=math10.index)
math11 = pd.Series(math11.math_score, index=math11.index)
math12 = pd.Series(math12.math_score, index=math12.index)

math_by_grade = pd.DataFrame({
    "9th": math9,
    "10th": math10,
    "11th": math11,
    "12th": math12
})

# Format Data
math_by_grade["9th"] = math_by_grade["9th"].map("{:,.2f}".format)
math_by_grade["10th"] = math_by_grade["10th"].map("{:,.2f}".format)
math_by_grade["11th"] = math_by_grade["11th"].map("{:,.2f}".format)
math_by_grade["12th"] = math_by_grade["12th"].map("{:,.2f}".format)

math_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# Starting with school summary since it has all data required
score_spend = school_summary[["Per Student Budget", "Average Math Score", "Average Reading Score", \
                              "Overall Average Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
# Adding the binning by spending column
score_spend["Spending Ranges (Per Student)"] = pd.cut(score_spend["Per Student Budget"], \
                                                         bins=spending_bins, labels=group_names)
# groupping by bins
score_spend = score_spend.groupby("Spending Ranges (Per Student)").mean()

# get rid of NOT required column
del score_spend["Per Student Budget"]

#Format data
score_spend["Average Math Score"] = score_spend["Average Math Score"].map("{:.2f}".format)
score_spend["Average Reading Score"] = score_spend["Average Reading Score"].map("{:.2f}".format)
score_spend["Overall Average Score"] = score_spend["Overall Average Score"].map("{:.2f}".format)
score_spend["% Passing Math"] = score_spend["% Passing Math"].map("{:.2f}%".format)
score_spend["% Passing Reading"] = score_spend["% Passing Reading"].map("{:.2f}%".format)
score_spend["% Overall Passing Rate"] = score_spend["% Overall Passing Rate"].map("{:.2f}%".format)

score_spend.head()
```
